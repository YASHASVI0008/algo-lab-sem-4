# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)

import matplotlib.pyplot as plt
import time

def even_odd_count(numbers):
    even_count = 0
    odd_count = 0
    for value in numbers:
        if value % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
    return even_count, odd_count

n_values = list(range(1, 1001, 10))
time_values = []

for n in n_values:
    data = list(range(n))

    start_time = time.perf_counter()
    even_odd_count(data)
    end_time = time.perf_counter()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (seconds)")
plt.title("O(n) Linear Time Complexity for Even/Odd Count")
plt.grid(True)
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithmic time. plot graph as well similar to above example.

import matplotlib.pyplot as plt
import time
import math

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

n_values = []
time_values = []

for i in range(1, 21):
    n_values.append(2**i)

for n in n_values:

    data = list(range(n))

    target = n - 1 if n > 0 else 0

    start_time = time.perf_counter()
    binary_search(data, target)
    end_time = time.perf_counter()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (seconds)")
plt.title("O(log n) Logarithmic Time Complexity for Binary Search")
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.

import matplotlib.pyplot as plt
import time
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    merged_list = []
    left_idx, right_idx = 0, 0

    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] < right[right_idx]:
            merged_list.append(left[left_idx])
            left_idx += 1
        else:
            merged_list.append(right[right_idx])
            right_idx += 1

    merged_list.extend(left[left_idx:])
    merged_list.extend(right[right_idx:])
    return merged_list

n_values = []
time_values = []


for i in range(1, 15):
    n_values.append(2**i * 100)

for n in n_values:

    data = [random.randint(0, n) for _ in range(n)]

    start_time = time.perf_counter()
    merge_sort(data)
    end_time = time.perf_counter()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (seconds)")
plt.title("O(n log n) Log-Linear Time Complexity for Merge Sort")
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example

import matplotlib.pyplot as plt
import time

def find_all_pairs(arr):
    count = 0
    for i in range(len(arr)):
        for j in range(len(arr)):

            count += 1
    return count

n_values = list(range(10, 101, 10))
time_values = []

for n in n_values:
    data = list(range(n))

    start_time = time.perf_counter()
    find_all_pairs(data)
    end_time = time.perf_counter()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (seconds)")
plt.title("O(n^2) Polynomial Time Complexity (Finding All Pairs)")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
# (Corrected to be Exponential Time Complexity, as per notebook flow)

import matplotlib.pyplot as plt
import time

def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

n_values = list(range(1, 30))
time_values = []

for n in n_values:
    start_time = time.perf_counter()
    fibonacci_recursive(n)
    end_time = time.perf_counter()

    time_values.append(end_time - start_time)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (seconds)")
plt.title("O(2^n) Exponential Time Complexity (Recursive Fibonacci)")
plt.grid(True)
plt.show()
```

```python
#Prob_8 plot graph similar to above figure

import matplotlib.pyplot as plt
import numpy as np


n_small = np.arange(1, 15)
n_medium = np.arange(1, 50)
n_large = np.arange(1, 1000)


def constant_time(n): return np.ones_like(n) * 2
def log_time(n): return np.log(n) * 10
def linear_time(n): return n
def n_log_n_time(n): return n * np.log(n)
def polynomial_time(n): return n**2
def exponential_time(n): return 2**n

plt.figure(figsize=(12, 8))


plt.plot(n_large, constant_time(n_large), label='O(1) Constant', linewidth=2)
plt.plot(n_large, log_time(n_large), label='O(log n) Logarithmic', linewidth=2)
plt.plot(n_large, linear_time(n_large), label='O(n) Linear', linewidth=2)
plt.plot(n_medium, n_log_n_time(n_medium), label='O(n log n) Log-Linear', linewidth=2)
plt.plot(n_medium, polynomial_time(n_medium) / 50, label='O(n^2) Polynomial (scaled)', linewidth=2)
plt.plot(n_small, exponential_time(n_small) / 100000, label='O(2^n) Exponential (scaled)', linewidth=2)

plt.xlabel("Input Size (n)")
plt.ylabel("Operations / Time (Theoretical)")
plt.title("Comparison of Time Complexities")
plt.legend()
plt.grid(True)
plt.ylim(0, 1000)
plt.xlim(0, 60)
plt.show()
```
