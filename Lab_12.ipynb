# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost

    return m

```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

# Exercise
1. Given p = [5, 10, 3, 12, 5, 50, 6]
2. Compute minimum cost
3. Print optimal parenthesization
4. Compare number of multiplications saved

```python
import sys

def matrix_chain_order(p):
    """
    Computes the minimum cost of multiplying a chain of matrices.
    p: List of dimensions where matrix Ai has dimensions p[i-1] x p[i]
    """
    n = len(p) - 1  # Number of matrices
    # m[i][j] stores the minimum number of scalar multiplications
    m = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
    # s[i][j] stores the index 'k' that achieved the optimal split
    s = [[0 for _ in range(n + 1)] for _ in range(n + 1)]

    # L is the chain length
    for L in range(2, n + 1):
        for i in range(1, n - L + 2):
            j = i + L - 1
            m[i][j] = sys.maxsize
            for k in range(i, j):
                # cost = multiplications in left + right + combining them
                cost = m[i][k] + m[k+1][j] + p[i-1] * p[k] * p[j]
                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k
    return m, s

def get_optimal_parens(s, i, j):
    """
    Recursively constructs the string representing the optimal parenthesization.
    """
    if i == j:
        return f"A{i}"
    else:
        return f"({get_optimal_parens(s, i, s[i][j])} x {get_optimal_parens(s, s[i][j] + 1, j)})"

def calculate_left_to_right_cost(p):
    """
    Calculates cost for a standard (non-optimized) left-to-right multiplication:
    ((((A1 x A2) x A3) x A4) ...)
    """
    cost = 0
    current_rows = p[0]
    current_cols = p[1]
    for i in range(2, len(p)):
        cost += current_rows * current_cols * p[i]
        current_cols = p[i]
    return cost

# --- Exercise Implementation ---

# 1. Given dimensions from Lab_12 exercise
p = [5, 10, 3, 12, 5, 50, 6]

# 2. Compute minimum cost and split table
m_table, s_table = matrix_chain_order(p)
min_cost = m_table[1][len(p)-1]

# 3. Get optimal parenthesization
optimal_order = get_optimal_parens(s_table, 1, len(p)-1)

# 4. Compare with standard order to find multiplications saved
standard_cost = calculate_left_to_right_cost(p)
savings = standard_cost - min_cost

# Output Results
print("--- Matrix Chain Multiplication Results ---")
print(f"Dimensions: {p}")
print(f"Minimum Multiplication Cost: {min_cost}")
print(f"Optimal Parenthesization: {optimal_order}")
print(f"Standard (Left-to-Right) Cost: {standard_cost}")
print(f"Total Multiplications Saved: {savings}")

print("\n--- DP Table (m) ---")
for row in m_table[1:]:
    print([format(val, '4d') if val != 0 else '   0' for val in row[1:]])
```
