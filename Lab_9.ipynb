# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

Now, let's implement
1. Modify program for user input graph using dictionary.
2. Display shortest path with total distance.
3. Add two more nodes in graph.
4. Test algorithm with different source nodes.

```python
def dijkstra_with_path(graph, start):
    if start not in graph:
        raise ValueError(f"Start node '{start}' not found in graph.")
    distances = {node: float('inf') for node in graph}
    previous = {node: None for node in graph}
    distances[start] = 0
    pq = [(0, start)]
    while pq:
        current_distance, current_node = heapq.heappop(pq)
        if current_distance > distances[current_node]:
            continue
        for neighbor, weight in graph[current_node].items():
            if weight < 0:
                raise ValueError("Dijkstra does not support negative edge weights.")
            new_distance = current_distance + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous[neighbor] = current_node
                heapq.heappush(pq, (new_distance, neighbor))
    return distances, previous

def build_path(previous, start, end):
    path = []
    current = end
    while current is not None:
        path.append(current)
        current = previous[current]
    path.reverse()
    if path and path[0] == start:
        return path
    return []

def print_shortest_paths(graph, source):
    distances, previous = dijkstra_with_path(graph, source)
    print(f"\nShortest paths from source node '{source}':")
    for node in graph:
        if distances[node] == float('inf'):
            print(f"{source} -> {node}: No path")
        else:
            path = build_path(previous, source, node)
            print(f"{source} -> {node}: Distance = {distances[node]}, Path = {' -> '.join(path)}")

def input_graph():

    n = int(input("Enter number of nodes: "))
    nodes = input(f"Enter {n} node names (space-separated): ").split()
    if len(nodes) != n:
        raise ValueError("Number of node names does not match.")
    graph = {node: {} for node in nodes}
    m = int(input("Enter number of edges: "))
    print("Enter edges in format: u v w")
    print("(For undirected graph, one entry will add both directions.)")
    for _ in range(m):
        u, v, w = input().split()
        w = int(w)
        if u not in graph or v not in graph:
            raise ValueError("Edge contains unknown node.")
        if w < 0:
            raise ValueError("Dijkstra requires non-negative weights.")
        graph[u][v] = w
        graph[v][u] = w
    return graph

graph = input_graph()

print("\nGraph dictionary:")
for node in graph:
    print(node, ":", graph[node])

k = int(input("\nHow many source nodes do you want to test? "))
sources = input(f"Enter {k} source node(s), space-separated: ").split()

if len(sources) != k:
    raise ValueError("Number of source nodes does not match.")

for src in sources:
    print_shortest_paths(graph, src)
```
